```python
state_data = pd.DataFrame({
    'State': ['CA', 'TX', 'FL','OH'],
    'Customer_ID': [25,50,75,100]
})
```

```python
m = folium.Map(location=[37.0902, -95.7129], zoom_start=4)
```

```python
# Create a Choropleth map
folium.Choropleth(
    geo_data=url, 
    name='choropleth',
    data=state_data,
    columns=['State', 'Customer_ID'],
    key_on='feature.id',  # Key to match GeoJSON with DataFrame
    fill_color='YlGn', 
    fill_opacity=0.7, 
    line_opacity=0.2,
    legend_name='Number of Customers by State'
).add_to(m)

# Save to an HTML file
m.save('us_customers_map.html')
```

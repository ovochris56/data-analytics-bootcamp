```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file and specify the sheet name
file_path = 'C:/Users/cfont/OneDrive/Desktop/christopherfontes/Documents/FontesCapstone_Dataset.xlsx' 
sheet_name = 'customer_conversion_dataset ' 
device_data = pd.read_excel(file_path, sheet_name=sheet_name)

# Group by 'DeviceType' and sum up the 'Conversion' values
conversion_per_device = device_data.groupby('DeviceType')['Conversion (Target)'].sum()

# Plot the grouped data
plt.figure(figsize=(12, 8))
conversion_per_device.plot(kind='bar', color='hotpink', edgecolor='black')
plt.title("Conversion per Device Type")
plt.xlabel("DeviceType")
plt.ylabel("Conversion (Target)")
plt.xticks(rotation=45)
plt.show()
```

```python
# Group by 'DeviceType' and sum up the 'Conversion' values
conversion_per_device = device_data.groupby('Location')['Conversion (Target)'].sum()

# Plot the grouped data
plt.figure(figsize=(12, 8))
conversion_per_device.plot(kind='bar', color='lightcoral', edgecolor='black')
plt.title("Conversion per Location")
plt.xlabel("Location")
plt.ylabel("Conversion (Target)")
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'C:/Users/cfont/OneDrive/Desktop/christopherfontes/Documents/FontesCapstone_Dataset.xlsx'
sheet_name = 'customer_conversion_dataset '
device_data = pd.read_excel('C:/Users/cfont/OneDrive/Desktop/christopherfontes/Documents/FontesCapstone_Dataset.xlsx', sheet_name='customer_conversion_dataset ')

# Calculate the average time spent per DeviceType
average_time_spent = device_data.groupby('DeviceType')['TimeSpent (minutes)'].mean()

# Plot the average time spent as a bar chart
plt.figure(figsize=(12, 8))
average_time_spent.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Average Time Spent per Device Type")
plt.xlabel("Device Type")
plt.ylabel("TimeSpent (minutes)")
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate the average time spent per DeviceType
average_time_spent = device_data.groupby('Conversion (Target)')['TimeSpent (minutes)'].mean()

# Plot the average time spent as a bar chart
plt.figure(figsize=(12, 8))
average_time_spent.plot(kind='bar', color='turquoise', edgecolor='black')
plt.title("Average Time Spent per Converted Customer")
plt.xlabel("Conversion (Target) ")
plt.ylabel("TimeSpent (minutes)")
plt.xticks(rotation=0)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the data
file_path = 'C:/Users/cfont/OneDrive/Desktop/christopherfontes/Documents/FontesCapstone_Dataset.xlsx'
sheet_name = 'customer_conversion_dataset '  
data = pd.read_excel(file_path, sheet_name=sheet_name)

# Group data by 'Age' and calculate conversion rate
conversion_data = data.groupby('Age')['Conversion (Target)'].mean() * 100  # Conversion rate as percentage

# Sort by highest conversion rate
conversion_data = conversion_data.sort_values(ascending=False)

# Plotting the bar chart
plt.figure(figsize=(12, 8))
plt.bar(conversion_data.index, conversion_data.values, color='teal', edgecolor='black')
plt.title("Customer Age Conversion Rate")
plt.xlabel("Age")
plt.ylabel("Conversion (Target)")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import pandas as pd
from scipy import stats  # For calculating the mode

# Load CSV file
file_path = 'StudentScores.csv'  # Replace with your CSV file path
df = pd.read_csv(file_path)

# Replace 'column_name' with the name of the column you want to analyze
column_name = 'math score'

# Check if the column exists and contains non-null values
if column_name in df.columns and df['math score].notnull().all():
    # Calculate the Mean (only if the column is numeric)
    if pd.api.types.is_numeric_dtype(df['math score']):
        mean_value = df['math score'].mean()
        median_value = df['math score'].median()

        # Calculate the Mode (handles both numeric and categorical data)
        mode_result = stats.mode(df['math score'], nan_policy='omit')
        mode_value = mode_result.mode[0] if len(mode_result.mode) > 0 else None

        # Print the results
        print(f"Mean: {mean_value}")
        print(f"Median: {median_value}")
        print(f"Mode: {mode_value}")
    else:
        print(f"Column '{'math score'}' is not numeric, skipping mean and median.")
else:
    print(f"Column '{'math score'}' is missing or contains only NaN values.")
```

```python
import pandas as pd
import numpy as np

# Load dataset (replace 'your_dataset.csv' with your file path)
file_path = 'StudentScores.csv'
df = pd.read_csv(file_path)

# Replace 'column_name' with the name of the column you're analyzing
column_name = 'math score'

# Check if the column exists and contains non-null values
if column_name in df.columns and df['math score'].notnull().all():
    
    # Calculate the Range (max - min)
    range_value = df['math score'].max() - df['math score'].min()

    # Calculate the Variance
    variance_value = np.var(df['math score'])

    # Calculate the Standard Deviation
    std_dev_value = np.std(df['math score'])

    # Print the results
    print(f"Range: {range_value}")
    print(f"Variance: {variance_value}")
    print(f"Standard Deviation: {std_dev_value}")
else:
    print(f"Column '{'math score'}' is missing or contains NaN values.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset (replace with your file path)
file_path = 'StudentScores.csv'
df = pd.read_csv(file_path)

# Clean data
df = df.dropna(subset=['math score', 'reading score'])
df = df[~df[['math score', 'reading score']].isin([float('inf'), -float('inf')]).any(axis=1)]

# Plot
plt.figure(figsize=(8, 6))
plt.scatter(df['math score'], df['reading score'], alpha=0.5, edgecolors='w')
plt.title('Students Scores')
plt.xlabel('math Score')
plt.ylabel('reading Score')

# Set axis limits
plt.xlim(df['math score'].min() - 5, df['math score'].max() + 5)
plt.ylim(df['reading score'].min() - 5, df['reading score'].max() + 5)

plt.grid(True)
plt.show()
```
